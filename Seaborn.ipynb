The Python visualization library Seaborn is based on matplotlib and provides a high-level interface for drawing attractive statistical graphics. More can be found at [seaborn.org](https://seaborn.pydata.org/)


```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

%matplotlib inline

# Load the data
tips = sns.load_dataset("tips")

sns.set_style("whitegrid")

g = sns.lmplot(x="tip",
                y ="total_bill",
              data = tips,
              aspect = 2)

g = (g.set_axis_labels("Tip", "Total bill (USD)"). 
     set(xlim=(0,10),ylim=(0,100)))

plt.title("Hello World")

# Show the plot
plt.show()
```

```python
# importing some more data
uniform_data = np.random.rand(10,12)
data = pd.DataFrame({'x':np.arange(1,101), 'y':np.random.normal(0,4,100)})

# importing some built-in datasets
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

Subplot grid for plotting conditional relationships

```python
g = sns.FacetGrid(titanic,col="survived",row="sex")
g = g.map(plt.hist,"age")

sns.factorplot(x="pclass",y="survived", hue="sex", data=titanic)
```

```python
sns.lmplot(x="sepal_width",y="sepal_length", hue="species",data=iris)
```

### Categorical Plots


```python
# Scatterplot with one categorical variable
sns.stripplot(x="species", y="petal_length",data=iris)
```

```python
# space it out a little bit more
sns.swarmplot(x="species", y="petal_length", data=iris)
```

```python
# barplot
sns.barplot(x="sex", y="survived", hue="class", data=titanic)
```

```python
# count plot
sns.countplot(x="deck", data=titanic, palette="Greens_d")
```

```python
# point plot
sns.pointplot(x="class",
             y = "survived",
             hue = "sex",
             data = titanic,
             palette={"male":"g",
                     "female":"m"},
             markers=["^","*"],
             linestyles=["-","--"])
```

```python
# boxplot
sns.boxplot(x="alive",
           y="age",
           hue="adult_male",
           data = titanic)
```

```python
# violin plot
sns.violinplot(x="age",
              y="sex",
              hue="survived",
              data=titanic)
```

```python
# Regression plots
f, ax = plt.subplots(figsize=(5,6))
sns.regplot(x="sepal_width",
           y="sepal_length",
           data=iris,
           ax=ax)
```

```python
# distribution plot 
plot = sns.distplot(data.y, kde=False, color="b")
```

```python
# heatmap
sns.heatmap(uniform_data,vmin=0,vmax=1)
```

```python
sns.heatmap(uniform_data,vmin=0,vmax=1, cmap="Blues")
```
